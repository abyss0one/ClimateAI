```python
import pandas as pd
```

```python
folder = "./datasets_models/"
```

```python
df = pd.read_csv(folder + "mean_data_x_day.csv")
```

```python
df
```

```python
df = df.sort_values(['comarca', 'data_lectura'])
df
```

```python
df['temp_anterior'] = df.groupby('comarca')['temp'].shift(1)
df['precip_anterior'] = df.groupby('comarca')['precip'].shift(1)
df['hume_anterior'] = df.groupby('comarca')['hume'].shift(1)
```

```python
df.to_csv(folder + "data.csv", index=False)
```
